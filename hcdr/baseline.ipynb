```python
app_train_origin =  pd.read_csv('./home-credit-default-risk/application_train.csv')
app_test_origin =  pd.read_csv('./home-credit-default-risk/application_test.csv')
app_train_origin.head()
```

```python
app_train = app_train_origin.copy()
app_test = app_test_origin.copy()
app_train.head()
```

```python
x_train = app_train.copy().drop(columns=['TARGET', 'SK_ID_CURR'])
y_train = app_train['TARGET'].copy()
id_train = app_train[['SK_ID_CURR']].copy()
x_train
```

```python
for col in x_train.columns:
    if x_train[col].dtype == "O":
        x_train[col] = x_train[col].astype("category")
```

```python
print("mean: {:.4f}".format(y_train.mean()))
y_train.value_counts()
```

```python
cv = list(StratifiedKFold(n_splits=5, shuffle=True, random_state=123).split(x_train, y_train))

# indexの確認:fold=0の学習データ
print("index(train):", cv[0][0])

# indexの確認:fold=0の検証データ
print("index(valid):", cv[0][1])
```

```python
# foldごとのindexリスト作成
cv = list(StratifiedKFold(n_splits=5, shuffle=True, random_state=123).split(x_train, y_train))

# 0fold目のindexのリスト取得
nfold = 0
idx_tr, idx_va = cv[nfold][0], cv[nfold][1]

# 学習データと検証データに分離
x_tr, y_tr, id_tr = x_train.loc[idx_tr, :], y_train[idx_tr], id_train.loc[idx_tr, :]
x_va, y_va, id_va = x_train.loc[idx_va, :], y_train[idx_va], id_train.loc[idx_va, :]
print(x_tr.shape, y_tr.shape, id_tr.shape)
print(x_va.shape, y_va.shape, id_va.shape)
```

```python
params = {
    'boostingtype': 'gbdt'
    ,'objective': 'binary'
    ,'metric': 'auc'
    ,'lerning_rate': 0.05
    ,'num_leaves': 32
    ,'n_estimators': 100000
    ,'random_state':123
    ,'importance_type': 'gain'
}

# モデルの学習
model = lgb.LGBMClassifier(**params)
model.fit(x_tr
          ,y_tr
          ,eval_set=[(x_tr, y_tr), (x_va, y_va)]
          ,early_stopping_rounds=100
          ,verbose=100
)

# モデルの保存
with open("model_lgb_fold0.pickle", "wb") as f:
    pickle.dump(model, f, protocol=4)
```

```python
# 学習データの推論地取得とROC計算
y_tr_pred = model.predict_proba(x_tr)[:, 1]
metric_tr = roc_auc_score(y_tr, y_tr_pred)

# 検証データの推論地取得とROC計算
y_va_pred = model.predict_proba(x_va)[:, 1]
metric_va = roc_auc_score(y_va, y_va_pred)

# 評価値を入れる変数の作成
metrics = []

# 評価値を格納
metrics.append([nfold, metric_tr, metric_va])

# 結果の表示
print("[auc] tr:{:.4f}, va:{:.4}".format(metric_tr, metric_va))
```

```python
# OOFを入れる変数の定義
train_oof = np.zeros(len(x_train))

# 検証データのindexに推論値を格納
train_oof[idx_va] = y_va_pred
```

```python
# 重要度の取得
imp_fold = pd.DataFrame({"col":x_train.columns, "imp":model.feature_importances_, "nfold":nfold})

# 確認(重要度の上位10個)
display(imp_fold.sort_values("imp", ascending=False)[:10])

# 重要度を格納する5-fold用のデータフレームの作成
imp = pd.DataFrame()
# imp_foldを5fold用データフレームに結合
imp = pd.concat([imp, imp_fold])
```

```python
# リスト型をArray型に変更
metrics = np.array(metrics)
print(metrics)

# 学習/検証データの評価値の平均値と標準偏差を導出
print("[cv] tr:{:.4f}+-{:.4f}, va:{:.4f}+-{:.4f}".format(
    metrics[:,1].mean(), metrics[:,1].std(),
    metrics[:,2].mean(), metrics[:,2].std(),
))

# oofの評価値を算出
print("[oof]{:.4f}".format(
    roc_auc_score(y_train, train_oof)
))
```
